```python

# Implementation of the Binary neural network function


# activation functions

def sig(xx):
    return 1 / (1 +  np.exp(-1.0*xx))
def nonlinear(spin,xx):
    return np.where(np.reshape(spin,(1,-1,1))==1,sig(xx),np.maximum(xx,0))
    
#evaluation of the cost function given a one-layer neural network and a dataset (x,y)    
def costfunction(samples,d_out,x,y):
    
    # samples come as an array of zeros and ones
    Ns = samples.shape[0]
    Npa = samples.shape[1]
    
    
    samplesi = samples[:,0:Nparameters-2]
    encoding = samples[:,Nparameters-2:]
    
    
    # Testing on different neural network encodings     
    samplesi = np.where(np.all(encoding == [0, 0], axis=1, keepdims=True), samplesi, samplesi)     #  0,1
    samplesi = np.where(np.all(encoding == [0, 1], axis=1, keepdims=True), 2*samplesi-1, samplesi) # -1,1
    samplesi = np.where(np.all(encoding == [1, 0], axis=1, keepdims=True), 3*samplesi-2, samplesi) # -2,1
    samplesi = np.where(np.all(encoding == [1, 1], axis=1, keepdims=True), 4*samplesi-3, samplesi) # -3,1 
    
    d_in = x.shape[1]
    Ntrain = x.shape[0]
    
    c = d_in*d_out

    # weights and bias
    
    W1 = np.reshape(samplesi[:,0:c],[Ns,d_in,d_out])
    b_1 = np.reshape(samplesi[:,c:c+d_out],[Ns,d_out])


    # binary neural network forward pass 

    p = sig(np.einsum('ij,kjl ->ikl',x,W1) + b_1) 
    Jw = -np.einsum('i,iwk ->wk', y,np.log(p))/Ntrain - np.einsum('i,iwk ->wk', 1-y,np.log(1-p))/Ntrain
    Jw = np.reshape(Jw,(-1))

    return  Jw


# neural network output given input x

def neural(samples,d_out,x,y):
    samplesi = np.copy(samples)
    samplesi = 2*samplesi-1
    d_in = x.shape[1]
    Ntrain = x.shape[0]
    
    c = d_in*d_out
    Ns = samples.shape[0]
    
    # forward pass
    W1 = np.reshape(samplesi[:,0:c],[Ns,d_in,d_out])
    b_1 = np.reshape(samplesi[:,c:c+d_out],[Ns,d_out])

    p = sig(np.einsum('ij,kjl ->ikl',x,W1) + b_1) 

    return  p


```

```python
%%time
#*********** brute force finding the neural network with the lowest cost function **************

# all the possible "neural nets"
bits = (np.array(list(it.product(range(2), repeat = Nparameters )),dtype=np.int8))
a = (2*(np.array(list(it.product(range(2), repeat = Nparameters )),dtype=np.int8)) - 1)


# first neural network evaluation
loss = costfunction(np.reshape(bits[0],(1,-1)),d_out,x_train_small,y_train)
Weights = bits[0]
print(bits[0],loss)
C = []
C.append(loss)

i=0
loss_new = costfunction(np.reshape(bits[i],(1,-1)),d_out,x_train_small,y_train)

# Loop over the rest of the entire set of possible BiNNs 

for i in range(1,bits.shape[0]):
 
    loss_new = costfunction(np.reshape(bits[i],(1,-1)),d_out,x_train_small,y_train) 
    C.append(loss_new)

    if loss_new<loss:

        loss = loss_new
        Weights = bits[i]
    
C = np.array(C)        
        

print(loss,Weights)
plt.plot(C)
plt.xlabel("configuration index")
plt.ylabel("Cost function")

# saving the full enumeration of the cost function C to a file. 
# This file is then loaded by the julia circuit optimizer mps_mpo_entangled_nn.jl 

np.savetxt("Cop_"+str(lx)+"x"+str(lx)+".txt",C)
```

```python
# accuracy of classification 

def class_accuracy(samples,d_out,x,y):
    samplesi = np.copy(samples)
    #samplesi = 2*samplesi-1
    d_in = x.shape[1]

    Ntra = x.shape[0]
    
    c = d_in*d_out
    Ns = samples.shape[0]
   
    W1 = np.reshape(samplesi[:,0:c],[Ns,d_in,d_out])
    b_1 = np.reshape(samplesi[:,c:c+d_out],[Ns,d_out])

    # binary 
    

    p = sig(np.einsum('ij,kjl ->ikl',x,W1) + b_1)

    out = np.where(p >0.5, 1, 0)
   

    out = np.reshape(out, (-1))
    a = np.abs(out-y)

    
    
    acc = 1-np.sum(a)/Ntra 
    acc = np.reshape(acc,(-1))

    return  acc 

# testing the accuracy of a few solutions  and a random neural network

sols = Weights
print(sols)


random_nn = np.random.binomial(1,0.5,(1,Nparameters))

acc_test = class_accuracy(np.reshape(sols,(1,-1)),d_out,x_test_small,y_test)
acc_train = class_accuracy(np.reshape(sols,(1,-1)),d_out,x_train_small,y_train)
print(acc_test,acc_train,y_test.shape)

acc_test = class_accuracy(random_nn,d_out,x_test_small,y_test)
acc_train = class_accuracy(random_nn,d_out,x_train_small,y_train)
print(acc_test,acc_train,y_test.shape)
```
